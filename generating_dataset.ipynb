```python
def get_rand_symbol():
    return chr(rnd.randint(33, 126))

INSERT_PROB = 0.01
def insert_symbol(word):
    if rnd.random() < INSERT_PROB:
        index = rnd.randint(0, len(word) - 1)
        return word[:index] + get_rand_symbol() + word[index:]
    else:
        return word

DELETE_PROB = 0.01
def delete_symbol(word):
    if rnd.random() < DELETE_PROB:
        index = rnd.randint(0, len(word) - 1)
        return word[:index] + word[index + 1:]
    else:
        return word

REPLACE_PROB = 0.02
def replace_symbol(word):
    if rnd.random() < REPLACE_PROB:
        index = rnd.randint(0, len(word) - 1)
        return word[:index] + get_rand_symbol() + word[index + 1:]
    else:
        return word

CHANGES_COUNT = 3
def noize_word(word):
    result = word

    for i in range(CHANGES_COUNT):
        result = insert_symbol(result)
        result = replace_symbol(result)
        result = delete_symbol(result)
    return result

ABBR_PROB = 0.4
DOT_PROB = 0.7
def abbr_word(word):
    if rnd.random() < ABBR_PROB:
        i = rnd.randint(3, 7)
        i = min(i, len(word))
        result = word[:i]
        if rnd.random() < DOT_PROB:
            result = result + '.'
        return result
    else:
        return word

def make_noize(words, with_abbr = True):
    return [abbr_word(noize_word(w)) for w in words]
```

```python
def gen_rand_word(len, lt):
    ans = []
    for _ in range(len):
        if (r2()):
            ans.append(random.choice(lt))
        else:
            ans.append(random.choice(NUMBERS))
    return ''.join(ans)
```

```python
def rand_space(x):
    
    if r2() and r2()and r2() and r2()and len(x) > 3:
        i = random.randint(1, len(x)-2)
        x = x[:i] + ' ' + x[i:]
        if r2() and r2():
            i = random.randint(1, len(x)-2)
            x = x[:i] + ' ' + x[i:]
            if r2():
                i = random.randint(1, len(x)-2)
                x = x[:i] + ' ' + x[i:]
    return x
```

```python
class GoodModelGenerator:
    def __init__(self):
        self._gen = self.generator()
        
    def generator(self):
        while True:
            ans = ''
            ln = random.randint(1, 6)

            w = gen_rand_word(ln, random.choice([LETTERS, LETTERS2]))
            
            if (r2() and r2() and r2()):
                w = gen_rand_word(random.randint(0, 2), LETTERS3) + \
                    gen_rand_word(random.randint(0, 3), random.choice([LETTERS, LETTERS2])) + \
                    gen_rand_word(random.randint(1, 7), NUMBERS)
            
            w = rand_brace(w, BRA0, BRA1)
            w = rand_space(w)
            
            yield w, TOKEN_MODEL
    
    def gen_name(self):
        return next(self._gen)
```

```python
def collapse(x):
    text = []
    mask = ''
    target = []
    
    for part, token in x:
        for word in part.split():
            text.append(word)
            mask = mask + token
            if token == TOKEN_MODEL:
                target.append(word)
    return {'text': ' '.join(text), 'mask': mask, 'target' : ' '.join(target)}
```

```python
def gen_dataset(N):
    
    gg = GetGood()
    ans = []
    for _ in range(N):
        ans.append(collapse(gg.gen_name()))
    
    ans = pd.DataFrame(ans)
    return ans
    
```
